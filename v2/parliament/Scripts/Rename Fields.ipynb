```python
import pymongo
import pprint
import json
import copy
import datetime
```

```python
config_file = open("../config.cfg")
config = json.load(config_file)
client = pymongo.MongoClient(config["mongodb_uri"])
db = client[config["database"]]
current_members = db["current_ls_members"]
```

```python
for row in current_members.find({})[5:]:
    terms = row["terms"]
    emails = row["email"]
    emails = [_ for _ in emails if _ != '']
    for i,term in enumerate(terms):
        terms[i]["party_id"] = terms[i].pop("party")
        terms[i]["constituency_id"] = terms[i].pop("constituency")
    old = copy.deepcopy(row)
    old["meta"]["old_record"] = ""
#     row["terms"] = terms
#     row["meta"]["old_record"] = old
#     row["email"] = emails
    pprint.pprint(row)
    current_members.update_one({"_id":row["_id"]},
                               {"$set":
                                {"terms":terms,
                                 "meta.old_record":old,
                                 "email":emails,
                                 "meta.fetched_on":datetime.datetime.now()}})
```
