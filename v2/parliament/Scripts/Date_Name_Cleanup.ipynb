```python
import pymongo
import json
from datetime import datetime
import pandas as pd
import re
import dateutil.parser as dparser
```

```python
config_file = open("../config.cfg")
config = json.load(config_file)
client = pymongo.MongoClient(config["mongodb_uri"])
db = client[config["database"]]
```

```python
def name_clean(member_name):
    name_list = member_name.split(',')
    if len(name_list) >= 2:
        name = name_list[1].strip()+" "+name_list[0].strip()
        name = name.replace("  "," ").replace('\xa0','')
    else:
        name = member_name
    return name
```

```python
def date_clean(dat):
    date_formats = ["%d.%m.%Y", "%d %b %Y", "%B %d, %Y", "%d-%m-%Y", "%Y-%m-%d", "%B,%Y", "%B, %Y", "%d %B %Y", "%d %B, %Y"]
    year_only_formats = ["%Y"]
    no_century = ["%d-%b-%y"]
    dat = dat.replace("b.","")
    dat = dat.replace("  "," ")
    dat = dat.strip()
    dat = dat.replace("st","").replace("nd","").replace("rd","").replace("th","")
    dat = dat.replace("Augu ","Aug")
    dat = dat.replace("Augu,","Aug")
    for date_format in date_formats:
        try:
            return datetime.strptime(dat,date_format).strftime("%Y-%m-%d"), True
        except:
            pass
    for date_format in year_only_formats:
        try:
            return datetime.strptime(dat,date_format).strftime("%Y"), True
        except:
            pass
    for date_format in no_century:
        try:
            return "19"+datetime.strptime(dat,date_format).strftime("%Y")+datetime.strptime(dat,date_format).strftime("-%m-%d"), True
        except:
            pass
#     months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
#     print re.findall(r"(?=("+'|'.join(string_lst)+r"))",x)
    try:
        return dparser.parse(dat,fuzzy=True).strftime("%Y-%m-%d"), True
    except:
        pass
    return dat, False
```
