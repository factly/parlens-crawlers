```python
import pymongo
import json
```

```python
config_file = open("../config.cfg")
config = json.load(config_file)
client = pymongo.MongoClient(config["mongodb_uri"])
db = client[config["database"]]
collection = db["rajya_sabha_archive_questions"]
members_collection = db["rs_all_members"]
```

```python
members = collection.distinct("members")
```

```python
not_found = []
multiple = []
for member in members:
    if members_collection.count_documents({"name":member}) == 0:
        not_found.append(member)
    elif members_collection.count_documents({"name":member}) > 1:
        multiple.append(member)
```

```python
names = list(members_collection.find({},{"name":1}))
```

```python
names = [_["name"] for _ in names]
```

```python
names2 = [_[_.index(" "):].replace(" ","").replace(".","").lower() for _ in names]
```
