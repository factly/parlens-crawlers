```python
import pandas as pd
import pymongo
import json
```

```python
config_file = open("../config.cfg")
config = json.load(config_file)
client = pymongo.MongoClient(config["mongodb_uri"])
db = client[config["database"]]
```

```python
collection = db["current_ls_members_updated"]
```

```python
df = pd.DataFrame(list(collection.find({})))
```

```python
df_temp = df.loc[:10].copy()
```

```python
possible_professions = {
    "Businessperson" : "Businessperson",
    "Social" : "Social Worker",
    "Artist" : "Artist",
    "Political" : "Political Worker",
    "Educationist" : "Educationist"
}
```

```python
for i,row in df_temp.iterrows():
    df_temp.loc[i,"profession_mod"] = ['']
```

```python
df_temp = df_temp.fillna("Not Available")
```

```python
for i,row in df_temp.iterrows():
    profs = []
    for possible in possible_professions.keys():
        if possible in row["profession"]:
            profs.append(possible_professions[possible])
    df_temp.loc[i,"profession_mod"] = profs
```
