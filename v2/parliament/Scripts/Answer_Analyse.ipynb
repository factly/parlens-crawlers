```python
import pymongo
import json
import re
```

```python
config_file = open("../config.cfg")
config = json.load(config_file)
client = pymongo.MongoClient(config["mongodb_uri"])
db = client[config["database"]]
```

```python
questions = db["lok_sabha_current_questions"]
```

```python
count = 0
answers = []
for q in questions.find({}):
    count+=1
    print(count," Answer: ",q['qref'])
    answers.append(q['answer'])
    print(q['answer'].strip("ANSWER\n").split("\n")[0])
    print(q['answer'].strip("ANSWER\n").split("\n")[1])
    if count == 100:
        break
```
