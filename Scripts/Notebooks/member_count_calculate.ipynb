```python
member_count = {}
```

```python
for document in collection.find():
    for member in document['members']:
        if member not in member_count:
            member_count[member] = 1
        else:
            member_count[member]+=1
```

```python
json.dump(member_count,open("member_count_16.json","w"))
```
