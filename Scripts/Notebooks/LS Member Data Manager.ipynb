```python
import pandas as pd
import pymongo
import json
import numpy as np
```

```python
data = pd.read_excel("../../Resources/mp_track.xlsx")
```

```python
data["Session"] = "16"
data["Minister"] = False
data.loc[pd.isna(data["Attendance"]),"Minister"] = True
data.loc[data["Membership"] == "Nominated", "State"] = "Not Applicable"
data.loc[data["Membership"] == "Nominated", "Constituency"] = "Not Applicable"
data.loc[pd.isna(data["State"]), "State"] = data[pd.isna(data["State"])]["Constituency"]
data = data.fillna("Not Available")
```

```python
reformatted_data = []
for i,row in data.iterrows():
    reformatted_data.append(json.loads(row.to_json()))
    reformatted_data[i]["ID"] = "ls_"+str(reformatted_data[i]["ID"])
```
